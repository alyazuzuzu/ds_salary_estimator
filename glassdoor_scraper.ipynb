```python
from selenium import webdriver
from shutil import which
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException
import pandas as pd
import time
```

```python
def jobs(keyword, num_pages):
    options = Options()
    options.add_argument("window-size=1920,1080")
    
    #Enter the chromedriver.exe path
    chrome_path = r"C:\Users\alya\Alya\ds_salary_estimator\chromedriver.exe"
    driver = webdriver.Chrome(executable_path=chrome_path, options=options)
    driver.get("https://www.glassdoor.com/Job/Home/recentActivity.htm")
    
    #Search for keyword
    search_input = driver.find_element(By.ID, "sc.keyword")
    search_input.send_keys(keyword)
    search_input.send_keys(Keys.ENTER)
    time.sleep(2)
    
    
    #Create Lists
    company_name = []
    rating = []
    job_title = []
    salary_est = []
    location = []
    job_description = []
    salary_estimate = []
    company_size = []
    company_type = []
    company_sector = []
    company_industry = []
    company_founded = []
    company_revenue = []
    
    
    
    #Set current page to 1
    current_page = 1     
        
    time.sleep(3)
    
    
    while current_page <= num_pages: #if true, current page will changed to the next page
        
        done = False
        while not done:
            job_cards = driver.find_elements(By.XPATH, "//article[@id='MainCol']//ul/li[@data-adv-type='GENERAL']")
            for card in job_cards:
                card.click()
                time.sleep(2)
                
                
                #Closes the signup prompt
                try:
                    driver.find_element(By.XPATH, ".//span[@class='SVGInline modal_closeIcon']").click()
                    time.sleep(2)
                except NoSuchElementException:
                    time.sleep(2)
                    pass

                
                #Click Show More to expand the Job Description
                try:
                    driver.find_element(By.XPATH, "//div[@class='css-t3xrds e856ufb2']").click()
                    time.sleep(1)
                except NoSuchElementException:
                    card.click()
                    print(str(current_page) + '#ERROR: no such element')
                    time.sleep(30)
                    driver.find_element(By.XPATH, "//div[@class='css-t3xrds e856ufb2']").click()
                    time.sleep(3)
                except ElementNotInteractableException:
                    card.click()
                    driver.implicitly_wait(30)
                    print(str(current_page) + '#ERROR: not interactable')
                    driver.find_element(By.XPATH, "//div[@class='css-t3xrds e856ufb2']").click()

                    
                #Scrape through the job information
                try:
                    company_name.append(driver.find_element(By.XPATH, "//div[@class='css-xuk5ye e1tk4kwz5']").text)
                except:
                    company_name.append("#N/A") #Set a not found value
                    pass
                
                try:
                    rating.append(driver.find_element(By.XPATH, "//*[@id='JDCol']/div/article/div/div[1]/div/div/div/div/div[1]/div[1]/span").text)
                except:
                    rating.append("#N/A") #Set a not found value

                try:
                    job_title.append(driver.find_element(By.XPATH, "//div[@class='css-1j389vi e1tk4kwz2']").text)
                except:
                    job_title.append("#N/A") #Set a not found value
                    pass

                try:
                    location.append(driver.find_element(By.XPATH, "//div[@class='css-56kyx5 e1tk4kwz1']").text)
                except:
                    location.append("#N/A") #Set a not found value
                    pass

                try:
                    job_description.append(driver.find_element(By.XPATH, "//div[@id='JobDescriptionContainer']").text)
                except:
                    job_description.append("#N/A") #Set a not found value
                    pass

                try:
                    salary_estimate.append(driver.find_element(By.XPATH, "//*[@id='JDCol']/div/article/div/div[1]/div/div/div/div/div[1]/div[4]").text)
                except:
                    salary_estimate.append("#N/A") #Set a not found value
                    pass
                
                try:
                    company_size.append(driver.find_element(By.XPATH, "//div[@id='CompanyContainer']//span[text()='Size']//following-sibling::*").text)
                except:
                    company_size.append("#N/A") #Set a not found value
                    pass
                
                try:
                    company_type.append(driver.find_element(By.XPATH, "//div[@id='CompanyContainer']//span[text()='Type']//following-sibling::*").text)
                except:
                    company_type.append("#N/A") #Set a not found value
                    pass
                    
                try:
                    company_sector.append(driver.find_element(By.XPATH, "//div[@id='CompanyContainer']//span[text()='Sector']//following-sibling::*").text)
                except:
                    company_sector.append("#N/A") #Set a not found value
                    pass
                    
                try:
                    company_industry.append(driver.find_element(By.XPATH, "//div[@id='CompanyContainer']//span[text()='Industry']//following-sibling::*").text)
                except:
                    company_industry.append("#N/A") #Set a not found value
                    pass
                    
                try:
                    company_founded.append(driver.find_element(By.XPATH, "//div[@id='CompanyContainer']//span[text()='Founded']//following-sibling::*").text)
                except:
                    company_founded.append("#N/A") #Set a not found value
                    pass
                    
                try:
                    company_revenue.append(driver.find_element(By.XPATH, "//div[@id='CompanyContainer']//span[text()='Revenue']//following-sibling::*").text)
                except:
                    company_revenue.append("#N/A") #Set a not found value
                    pass
                    
                    
                    
                done = True
                
       #Click the Next Page button         
        if done:
            print(str(current_page) + ' ' + 'out of' +' '+ str(num_pages) + ' ' + 'pages done')
            driver.find_element(By.XPATH, "//span[@alt='next-icon']").click()   
            current_page = current_page + 1
            time.sleep(5)
            



    #Converts the dictionary object into a pandas DataFrame.
    driver.close()
    df = pd.DataFrame({'Company Name': company_name, 
    'Rating' : rating,
    'Job Title': job_title,
    'Location': location,
    'Job Description': job_description,
    'Salary Estimate': salary_estimate,
    'Size': company_size,
    'Type of Ownership': company_type,
    'Sector': company_sector,
    'Industry' : company_industry,
    'Founded' : company_founded, 
    'Revenue': company_revenue})
    
    df.to_csv('Data Scientist 2.csv')
```
